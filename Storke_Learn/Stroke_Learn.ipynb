```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/data_2.csv',dtype=float)
```

```python
df.head()
```

```python
df.describe()
```

```python
#stroke 건수확인
print(df['stroke'].value_counts())
print('정상확률')
print(df['stroke'].value_counts()[0]/df['stroke'].value_counts().sum()*100)
```

```python
#결측치 확인
df.isna().sum()
```

```python
# 데이터 형 변환
integer = ['stroke','DM','current_ascvd','HTN_DIAG','gender','smk1','smk2','group1','group2']

for i in integer:
    df[i] = df[i].astype('int')
df.dtypes
```

```python
df.head()
```

```python
#스케일
from sklearn.preprocessing import MinMaxScaler
scaler = ['AGE','BMI','BP_HIGH','TOT_CHOLE','HDL_CHOLE','LDL_CHOLE','interval','econo','drinking']

min = MinMaxScaler()
min.fit(df[scaler])
df[scaler] = min.transform(df[scaler])

df.head()
```

```python
#1. 데이터 분류

from sklearn.model_selection import train_test_split

X = df.drop('stroke', axis=1)
Y = df['stroke']

X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size= 0.2,shuffle=True, random_state=1000) # stratify데이터 분포도 골고루하기

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier

model1 = RandomForestClassifier()
model1.fit(X_train,y_train)
pred1 = model1.predict(X_test)
print('Stroke Pred Count: ',pred1.sum())
```

```python
from sklearn.ensemble import AdaBoostClassifier

model2 = AdaBoostClassifier()
model2.fit(X_train,y_train)
pred2 = model2.predict(X_test)

print('Stroke Pred Count: ',pred2.sum())
```

```python
# 앙상블
from sklearn.ensemble import VotingClassifier
clf = VotingClassifier(estimators=[('rf',model1),('ad',model2)],voting='hard')
clf.fit(X_train,y_train)
pred3 = clf.predict(X_test)
print('Stroke Pred Count: ',pred3.sum())
```

```python
print('Stroke Count: ',y_test.sum())
```

```python
"""
평가지표 참고: https://han-py.tistory.com/363
"""
from sklearn.metrics import f1_score,recall_score,accuracy_score

print('-'*10+'accuracy_score'+'-'*10)
print('rf',accuracy_score(y_test,pred1))
print('Ada',accuracy_score(y_test,pred2))
print('Voting',accuracy_score(y_test,pred3))

print('-'*10+'recall_score'+'-'*10)
print('rf',recall_score(y_test,pred1))
print('Ada',recall_score(y_test,pred2))
print('Voting',recall_score(y_test,pred3))

print('-'*10+'f1_score'+'-'*10)
print('rf',f1_score(y_test,pred1))
print('Ada',f1_score(y_test,pred2))
print('Voting',f1_score(y_test,pred3))

print()
print('stroke 비율')
print(y_test.sum()/y_test.count() * 100,'%')
```

```python
from sklearn.metrics import plot_precision_recall_curve
fig = plt.figure()  # figure하나를 만들어 주자.
ax = fig.gca()  # gca를 사용해서 ax를 사용할 수 있다.
plot_precision_recall_curve(model1, X_test, y_test, ax=ax)
plot_precision_recall_curve(model2, X_test, y_test, ax=ax)
```

```python
from sklearn.metrics import plot_roc_curve
fig = plt.figure()
ax = fig.gca()
plot_roc_curve(model1, X_test, y_test, ax=ax)
plot_roc_curve(model2, X_test, y_test, ax=ax)
```
