```python
import tensorflow as tf
from tensorflow.keras import layers, Sequential
```

```python
help(layers.Conv2D)
#Conv2D(filters, kernel_size, strides=(1, 1), padding='valid', data_format=None, 
#dilation_rate=(1, 1), groups=1, activation=None, use_bias=True, kernel_initializer='glorot_uniform', 
#bias_initializer='zeros', kernel_regularizer=None, bias_regularizer=None, activity_regularizer=None, kernel_constraint=None, bias_constraint=None, **kwargs)
```

```python
#Conv2D 하나로 이루어진 모델
def build_model1(input_shape):
    model = layers.Conv2D(1, kernel_size = (3,3), strides = (1,1), padding="same",
                          activation="relu", 
                          input_shape=input_shape[1:])
    return model
```

```python
#Conv2D 두개로 이루어진 모델을 완성하세요 
def build_model2(input_shape):
    model = Sequential([layers.Conv2D(4, kernel_size = (3,3), strides = (1,1), padding="same", input_shape=input_shape[1:]),
                        layers.Conv2D(4, kernel_size = (3,3), strides = (1,1), padding="same")])
    return model
```

```python
#Conv2D 세개로 이루어진 모델을 완성하세요 
def build_model3(input_shape):
    model = Sequential()
    model.add(layers.Conv2D(2, kernel_size = (3,3), strides = (1,1), padding="same", input_shape=input_shape[1:]))
    model.add(layers.Conv2D(4, kernel_size = (3,3), strides = (1,1), padding="same"))
    model.add(layers.Conv2D(8, kernel_size = (3,3), strides = (1,1)))
    return model
```

```python
input_shape = (1, 5, 5, 1) #입력되는 데이터의 형식
#(batch_size, 이미지 가로, 이미지 세로, 이미지 채널)
model1 = build_model1(input_shape)
model2 = build_model2(input_shape)
model3 = build_model3(input_shape)

x = tf.ones(input_shape) #모든 값이 1인 임의의 데이터 생성
#print(x)
#output 형식
print("model1을 통과한 결과:", model1(x).shape,"필터 1개, 패딩을 same로 설정하여 입력과 동일하게 이미지 사이즈(가로,세로) 출력")
print("model2을 통과한 결과:", model2(x).shape,"필터 4개, 패딩을 same로 설정하여 입력과 동일하게 이미지 사이즈(가로,세로) 출력")
print("model3을 통과한 결과:", model3(x).shape,"필터 8개, 패딩을 same로 설정하여 입력과 동일하게 이미지 사이즈(가로,세로) 출력")
```
